```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
def courbe(data,liste):
    fig, ax = plt.subplots(figsize=(15, 5))
    for e in liste:
        top = data.loc[e,:]
        top_score = top.groupby(['Module','actor']).score.max()
        top_temps = top.groupby(['Module','actor']).temps.sum()
        
        df7 =pd.DataFrame({'Score':top_score,'Temps':top_temps})
        df7["Temps"] = np.round(df7['Temps']/60,2)
        df7.sort_values(by = 'Temps', ascending = False, inplace = True)
        
        x= df7['Temps']
        y= df7['Score']
        
        ax.plot(x,y, label = e)
        ax.set_xlabel('temps en min')
        ax.set_ylabel('score max')
        ax.set_title('Courbe des 10 top modules')
        ax.legend();
```

Pour tous les modules:
représenter l’ensemble par un nuage de points dans un même repère sur une même figure. A chaque apprenant correspond un point dont l’abscisse est le rang de celui-ci. L’ordonné est déterminé par le temps global sur le module.


```python
def courbe(data,liste):
    fig, ax = plt.subplots(figsize=(15, 5))
    for e in liste:
        top = data.loc[e,:]
        top_score = top.groupby(['Module','actor']).score.max()
        top_temps_global = top.groupby(['Module','actor']).temps.sum()        
        df9 = pd.DataFrame({'Score':top_score,'Temps global':top_temps_global})
        df9["Temps global"] = np.round(df9['Temps global']/60,2)
        df9.sort_values(by = 'Score', ascending = False, inplace = True)
        df9['Rang'] = list(range(1,len(df9['Score'])+1))
        
        x= df9['Rang']
        y= df9['Temps global']
        
        ax.scatter(x,y, s=80)
        ax.set_xlabel('Rang')
        ax.set_ylabel('Temps global')
        ax.set_title('Courbe par un nuage de points des modules');
```

Un camembert avec le taux des modules catégorisés en quatre classes: 
Classe A: taux de réussite > = 80%
Classe B: taux de réussite > = 60%
Classe C: taux de réussite > = 50%
Classe D: taux de réussite < 50%


```python
def camembert(df):
    classe_A = len(df.loc[(df['Taux de réussite'] >= 80)])
    classe_B = len(df.loc[(df['Taux de réussite'] < 80) & (df['Taux de réussite'] >= 60)])
    classe_D = len(df.loc[(df['Taux de réussite'] < 60) & (df['Taux de réussite'] >= 50)])
    classe_C = len(df.loc[(df['Taux de réussite'] < 50)])
    name=['ClasseA','ClasseB','ClasseD','ClasseC',]
    data1= [classe_A,classe_B,classe_D,classe_C]
    explode=(0, 0, 0, 0)
    plt.pie(data1, explode=explode, labels = name, autopct='%1.2f%%', startangle=50)
    plt.axis('equal')
    plt.show()
```
